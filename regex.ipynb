```python
import re
import pandas as pd
```

```python


# Define regex patterns for each level of Bloom's Taxonomy
bloom_patterns = {
    "Knowledge": r"\b(arrang|choos|cite|copi|defin|describ|draw|duplic|identifi|indic|label|list|match|memor|name|order|outlin|quot|read|recal|recit|recogn|record)\b",
    "Comprehension": r"\b(articul|associ|character|cite|clarifi|classifi|compar|contrast|convert|defend|demonstr|describ|differenti|discus|distinguish|estim|explain|express|extend|extrapol|gener|give|identifi)\b",
    "Application": r"\b(act|adapt|appli|back/back up|calcul|chang|choos|classifi|complet|comput|construct|demonstr|develop|discov|dramat|employ|experi|explain|gener|identifi|illustr|implement|interpret|interview)\b",
    "Analysis": r"\b(analyz|apprais|break|break down|calcul|categor|classifi|compar|conclud|contrast|correl|critic|debat|deduc|detect|diagnos|diagram|differenti|discov|discrimin|dissect|distinguish|divid|evalu)\b",
    "Evaluation": r"\b(apprais|argu|arrang|assess|attach|choos|compar|conclud|consid|core|counsel|creat|critiqu|decid|defend|describ|design|determin|discrimin|estim|evalu|explain|grade)\b",
    "Creation": r"\b(arrang|assembl|categor|choos|collect|combin|compil|compos|construct|creat|design|develop|devi|estim|evalu|explain|facilit|formul|gener|hypothes|improv|integr|invent)\b"
}

def categorize_question(question):
    for level, pattern in bloom_patterns.items():
        if re.search(pattern, question, re.IGNORECASE):
            return level
    return "Unknown"  # If no pattern matches

# Example usage
questions = [
    "Define photosynthesis.",
    "Explain the process of cellular respiration.",
    "Solve the equation 2x + 3 = 7.",
    "Compare and contrast mitosis and meiosis.",
    "Evaluate the impact of climate change on biodiversity.",
    "Design an experiment to test the effect of light on plant growth."
]

for q in questions:
    category = categorize_question(q)
    print(f"Question: {q}\nCategory: {category}\n")
```
